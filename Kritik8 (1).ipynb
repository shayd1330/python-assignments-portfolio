```python
def gradient_descent(x0, y0,f, grad_f, alpha, num_iterations):
    x, y = x0, y0 # initialize x and y with the initial points (x0 and y0)
    for i in range(num_iterations):
    #gradient of f at (x, y)
        grad_x, grad_y = grad_f(x, y)
    # Update x and y by taking a step in the opposite direction of the gradient
        x -= alpha*grad_x
        y -= alpha*grad_y
    #return the coordinates of the final point after the gradient descent
    return (x, y)

#define the function f1(x, y) = x**2 + y**2 and its gradient
def fun_1(x,y):
    return (x**2+y**2)

def grad_f_1(x,y):
    grad_x_1 = 2*x
    grad_y_1 = 2*y
    return (grad_x_1, grad_y_1)
        
#applying gradient descent to test cases for function 1
#test case 1 
(x0, y0) = (0.1, 0.1)
alpha = 0.1
num_iterations = 10 #max of 10
test_1 = gradient_descent(x0, y0,fun_1, grad_f_1, alpha, num_iterations)
print ("Function 1, test case 1 minimum=", (test_1))
    
#test case 2
(x0, y0) = (-1, 1)
alpha = 0.01
num_iterations = 100 #max 100
test_2 = gradient_descent(x0, y0,fun_1, grad_f_1, alpha, num_iterations)
print ("Function 1, test case 2 minimum=", (test_2))


#defining a function fun_2(x, y) and its gradient
import numpy as np
def fun_2(x, y):
    return (1-np.exp(-x**2-(y-2)**2)-2*np.exp(-x**2-(y+2)**2))
        
def grad_f_2(x, y):
    grad_x_2 = 2*x*np.exp(-x**2-(y-2)**2)+4*np.exp(-x**2-(y+2)**2)
    grad_y_2 = 2*(y-2)*np.exp(-x**2-(y-2)**2)+4*(y+2)*(np.exp(-x**2-(y+2)**2))    
    return (grad_x_2, grad_y_2)
    
#applying gradient descent to test cases for function 2
#test case 3
(x0, y0) = (0, 1)
alpha = 0.01
num_iterations = 10000 #max 10000
test_3 = gradient_descent(x0, y0,fun_2, grad_f_2, alpha, num_iterations)
print ("Function 2, test case 1 minimum=", (test_3))

#test case 4
(x0, y0) = (0, -1)
alpha = 0.01
num_iterations = 10000 #max 10000
test_4 = gradient_descent(x0, y0,fun_2, grad_f_2, alpha, num_iterations)
print ("Function 2, test case 2 minimum=", (test_4))

#plot function 2
from mpl_toolkits import mplot3d #for 3D plots
import matplotlib.pyplot as plt #usual matplotlib

X=np.linspace(-5,5,100)
Y=np.linspace(-5,5,100)
x,y=np.meshgrid(X,Y)
z= fun_2(x, y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")  
plt.show()
```
